```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1d

```python
data_1d_pbc = pd.read_csv('outputs/1d-pbc.csv', sep=',')
data_1d_apbc = pd.read_csv('outputs/1d-apbc.csv', sep=',')
```

```python
fig, axes = plt.subplots(2, 2, sharex='col', sharey='row', squeeze=True, figsize=(9,7), dpi=100)

for N in data_1d_pbc['N'].unique():
    data_1d_pbc_N = data_1d_pbc[data_1d_pbc['N'] == N].sort_values('J')
    axes[0,0].plot(data_1d_pbc_N['J'], data_1d_pbc_N['M'], 'o-', label='N = {:.0f}'.format(N))
    axes[1,0].plot(data_1d_pbc_N['J'], data_1d_pbc_N['E_0'], 'o-', label='N = {:.0f}'.format(N))

    data_1d_apbc_N = data_1d_apbc[data_1d_apbc['N'] == N].sort_values('J')
    axes[0,1].plot(data_1d_apbc_N['J'], data_1d_apbc_N['M'], 'o-', label='N = {:.0f}'.format(N))
    axes[1,1].plot(data_1d_apbc_N['J'], data_1d_apbc_N['E_0'], 'o-', label='N = {:.0f}'.format(N))

    axes[0,0].set_ylabel(r'$M$')
    axes[1,0].set_xlabel(r'$J$')
    axes[1,0].set_ylabel(r'$E_0$')
    axes[1,1].set_xlabel(r'$J$')
    axes[0,0].set_title(r'1d, PBC, $\epsilon = 1$')
    axes[0,1].set_title(r'1d, APBC, $\epsilon = 1$')
    axes[1,0].legend(loc='upper center', bbox_to_anchor=(1.1, -0.2), fancybox=True, shadow=False, ncol=3)

for ax in axes.reshape(-1):
    ax.grid()

fig.savefig('outputs/1d-pbc-apbc.png', dpi=150, bbox_inches='tight')
```

# 2d

```python
data_2d_pbc = pd.read_csv('outputs/2d-pbc.csv', sep=',')
#data_1d_apbc = pd.read_csv('outputs/1d-apbc.csv', sep=',')
```

```python
fig, axes = plt.subplots(2, 2, sharex='col', sharey='row', squeeze=True, figsize=(9,7), dpi=100)

for N_i, N_j in list(data_2d_pbc[['N_i', 'N_j']]
                     .drop_duplicates().itertuples(index=False, name=None)):
    data_2d_pbc_N = data_2d_pbc[(data_2d_pbc['N_i'] == N_i) &
                                (data_2d_pbc['N_j'] == N_j)].sort_values('J')
    axes[0,0].plot(data_2d_pbc_N['J'], data_2d_pbc_N['M'],
                   'o-', label=r'N = {:.0f} $\times$ {:.0f}'.format(N_i, N_j))
    axes[1,0].plot(data_2d_pbc_N['J'], data_2d_pbc_N['E_0'],
                   'o-', label=r'N = {:.0f} $\times$ {:.0f}'.format(N_i, N_j))

#    data_1d_apbc_N = data_1d_apbc[data_1d_apbc['N'] == N].sort_values('J')
#    axes[0,1].plot(data_1d_apbc_N['J'], data_1d_apbc_N['M'], 'o-', label='N = {:.0f}'.format(N))
#    axes[1,1].plot(data_1d_apbc_N['J'], data_1d_apbc_N['E_0'], 'o-', label='N = {:.0f}'.format(N))

    axes[0,0].set_ylabel(r'$M$')
    axes[1,0].set_xlabel(r'$J$')
    axes[1,0].set_ylabel(r'$E_0$')
    axes[1,1].set_xlabel(r'$J$')
    axes[0,0].set_title(r'2d, PBC, $\epsilon = 1$')
    axes[0,1].set_title(r'2d, APBC, $\epsilon = 1$')
    axes[1,0].legend(loc='upper center', bbox_to_anchor=(1.1, -0.2), fancybox=True, shadow=False, ncol=3)

for ax in axes.reshape(-1):
    ax.grid()

fig.savefig('outputs/2d-pbc-apbc.png', dpi=150, bbox_inches='tight')
```
